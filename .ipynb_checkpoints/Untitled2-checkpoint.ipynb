```python
# label several label with specific grading system.
def SC_LabelEncoder1(text):
    if text == "credit_card":
        return 1
    elif text == "car":
        return 2
    elif text == "small_business":
        return 3
    elif text == "other":
        return 4
    elif text == "wedding":
        return 5
    elif text == "debt_consolidation":
        return 6
    elif text == "home_improvement":
        return 7
    elif text == "major_purchase":
        return 8
    elif text == "medical":
        return 9
    elif text == "moving":
        return 10
    elif text == "vacation":
        return 11
    elif text == "house":
        return 12
    elif text == "renewable_energy":
        return 13
    else:
        return 0
    
df_loanfile["purpose"] = df_loanfile["purpose"].apply(SC_LabelEncoder1)
```
